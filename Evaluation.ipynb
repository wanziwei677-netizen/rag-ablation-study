```python
import os
import json
import nest_asyncio
import pandas as pd
from datasets import Dataset
import os
from dotenv import load_dotenv

load_dotenv()

from ragas import evaluate
from ragas.metrics import (
    context_recall,
    context_precision,
    faithfulness,
    answer_correctness
)

from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper

nest_asyncio.apply()

openai_api_key = os.getenv("OPENAI_API_KEY")

base_llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
base_embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

ragas_llm = LangchainLLMWrapper(base_llm)
ragas_emb = LangchainEmbeddingsWrapper(base_embeddings)

for metric in [context_recall, context_precision, faithfulness, answer_correctness]:
    metric.llm = ragas_llm
    if hasattr(metric, 'embeddings'):
        metric.embeddings = ragas_emb

file_path = "Rag_result_1.json" # change result_file name

with open(file_path, "r", encoding="utf-8") as f:
    results_data = json.load(f)

questions = [item["query"] for item in results_data]
ground_truths = [item["gold_answer"] for item in results_data]
answers = [item["generated_answer"] for item in results_data]
contexts = [[ctx["text"] for ctx in item.get("retrieved_contexts", [])] for item in results_data]

data_samples = {
    "question": questions,
    "answer": answers,
    "contexts": contexts,
    "ground_truth": ground_truths
}
dataset = Dataset.from_dict(data_samples)

result = evaluate(
    dataset=dataset,
    metrics=[
        context_recall,      
        context_precision,   
        faithfulness,        
        answer_correctness  
    ],
    llm=ragas_llm,
    embeddings=ragas_emb,
)

print(result)

df_result = result.to_pandas()
display(df_result)
```

```python
weights = {
    'answer_correctness': 0.4,
    'faithfulness': 0.3,
    'context_recall': 0.2,
    'context_precision': 0.1
}

df_result['RAG_Score'] = (
    df_result['answer_correctness'] * weights['answer_correctness'] +
    df_result['faithfulness'] * weights['faithfulness'] +
    df_result['context_recall'] * weights['context_recall'] +
    df_result['context_precision'] * weights['context_precision']
)

display(df_result) 

overall_rag_score = df_result['RAG_Score'].mean()
print(f"\n Current System's Overall RAG Score (North Star Metric): {overall_rag_score:.4f}")
```
